# Creating static maps in GeoPandas

```python
# Necessary imports
import geopandas as gpd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the data files
grid = gpd.read_file("Data/TravelTimes_to_5975375_RailwayStation.shp")
roads = gpd.read_file("Data/roads.shp")
metro = gpd.read_file("Data/metro.shp")
```

```python
# Re-project coordinates into the same CRS
gridCRS = grid.crs
roads['geometry'] = roads['geometry'].to_crs(crs=gridCRS)
metro['geometry'] = metro['geometry'].to_crs(crs=gridCRS)
```

```python
# Plot the map using GeoPandas/MPL
mapAx = grid.plot(column="car_r_t", linewidth=0.03, cmap="jet_r", scheme="quantiles", k=9, alpha=0.9)
roads.plot(ax=mapAx, color="grey", linewidth=1.5) # Add roads
metro.plot(ax=mapAx, color="red", linewidth=2.5) # Add metro
plt.tight_layout()
```
